```python
#http://www.worldportsource.com/ports/index/ALB.php



import urllib.request
import requests
from bs4 import BeautifulSoup
import urllib.request
import csv

url='http://www.worldportsource.com/ports/index/ALB.php'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
page=requests.get(url,headers=headers)
if page.status_code==200:
        print ("Site downloaded successfully")

```

```python
soup = BeautifulSoup(page.content, 'html.parser')
table=soup.find_all('table', class_='form')[0].find_all('a',href=True)
table
```

```python
Links=[table[i]['href'] for i in range(len(table)) ]
Links
```

```python
LinksCount=[Links[i] for i in range (len(Links)) if len(str(Links[i]))>10]
LinksCount
```

```python
LinksCountries=['http://www.worldportsource.com'+LinksCount[i] for i in range(len(LinksCount))]
```

```python
LinksCountries[0:2:1]
```

```python
import urllib.request
import requests
from bs4 import BeautifulSoup
import urllib.request
import csv

LinkPorts=[]
for kk in range (len(LinksCountries)):
    url=LinksCountries[kk]
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    page=requests.get(url,headers=headers)
    if page.status_code==200:
        print ("Site downloaded for "+url)
    soup = BeautifulSoup(page.content, 'html.parser')
    table=soup.find_all('table', class_='form')[0].find_all('a',href=True)
    Links=[table[i]['href'] for i in range(len(table)) ]
    LinksPorts_kk=[Links[i] for i in range (len(Links)) if len(str(Links[i]))>10]
    LinkPorts.extend(LinksPorts_kk)
```

```python
len(LinkPorts)
```

```python
LinkPortsFinal=['http://www.worldportsource.com'+LinkPorts[i] for i in range(len(LinkPorts))]
```

```python
LinkPortsFinal2=[LinkPortsFinal[i] for i in range(len(LinkPortsFinal)) if (LinkPortsFinal[i].split('http://www.worldportsource.com/ports/')[1].split('/')[0]!='review')]
```

```python
len(LinkPortsFinal2)
```

```python
from pandas import DataFrame as df
```

```python
LinksData=df(LinkPortsFinal2)
```
